```python
##### import os

while True:
    print("\n======= FILE HANDLING MENU ========")
    print("1. Create File")
    print("2. Write to File")
    print("3. Read to File")
    print("4. Append File")
    print("5. Replace Text in File")
    print("6. Rename File")
    print("7. Delete File")
    print("8. Exit")

    choice = input("Enter your choice (1-8): ")

    # Create file
    if choice == '1':
        name = input("Enter file name: ") + ".txt"
        f = open(name, "w")
        f.close()
        print("***File created successfully***!")

    # Write file
    elif choice == '2':
        name = input("Enter file name: ") + ".txt"
        text = input("Enter text to write: ")
        f = open(name, "w")
        f.write(text)
        f.close()
        print("***Text written successfully***!")

    # Read file
    elif choice == '3':
        name = input("Enter file name: ") + ".txt"
        if os.path.exists(name):
            f = open(name, "r")
            print("\nFile Contents:\n------------------")
            print(f.read())
            f.close()
        else:
            print("File not found!")

    # Append file
    elif choice == '4':
        name = input("Enter file name: ") + ".txt"
        if os.path.exists(name):
            text = input("Enter text to add: ")
            f = open(name, "a")
            f.write("\n" + text)
            f.close()
            print("***Text added successfully!***")
        else:
            print("File not found!")
    
    # Replace text
    elif choice == '5':
        name = input("Enter file name: ") + ".txt"
        if os.path.exists(name):
            old = input("Enter word to replace: ")
            new = input("Enter new word: ")
            f = open(name, "r")
            data = f.read()
            f.close()
            data = data.replace(old, new)
            f = open(name, "w")
            f.write(data)
            f.close()
            print("***Text replaced successfully***!")
        else:
            print("File not found!")

    # Rename file
    elif choice == '6':
        old = input("Enter old file name: ") + ".txt"
        new = input("Enter new file name: ") + ".txt"
        if os.path.exists(old):
            os.rename(old, new)
            print("***File renamed successfully***!")
        else:
            print("File not found!")

    # Delete file
    elif choice == '7':
        name = input("Enter file name to delete: ") + ".txt"
        if os.path.exists(name):
            os.remove(name)
            print("***File deleted successfully***!")
        else:
            print("File not found!")

    # Exit
    elif choice == '8':
        print("Exiting.......Thank you!!!")
        break

    else:
        print("XXXXX Invalid choice! Try again XXXXX")
```
